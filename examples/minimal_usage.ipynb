```python
import math
from randmut import randmut
```

```python
def fun(x):
    return sum([math.sin(v) for v in x])
```

```python
#minimizing the sum of three sinuses from -pi to pi
(x, f, _) = randmut(fun, 3 * [(-math.pi, math.pi)])
print("Minimal value found:", f)
print("With parameters:", x)
```

```python
#doing the same silently
(x, f, _) = randmut(fun, 3 * [(-math.pi, math.pi)], disp=False)
print("Minimal value found:", f)
print("With parameters:", x)
```
